8. Complete the following assignment. Fisher's Tea Experiment

In Fisher's original experiment, Dr. Bristol claimed to differentiate between cups where tea or milk was poured first. Fisher devised a hypothesis test to see if she could consistently perform better than random guessing. Similarly, our experiment involves STA130 students, with the goal of assessing if their ability to identify the pouring order is due to skill or chance.

Null Hypothesis(H0): The proportion of students who can correctly identify the pouring order is 0.5, implying random guessing.
Alternative Hypothesis(H1): The proportion of students who can correctly identify the pouring order is greater than 0.5, indicating some level of skill.

we will use bootstrapping to simulate the sampling distribution of the sample proportion. For each romdomly selected resample, calculate the proportion of "correct guesses and build a distribution of these proportions.
Use this distribution to calculate the confidence interval and p-value based on the proportion of resamples that are greater than or equal to the observed proportion of 0.6125(49/80).



```python
import numpy as np
import matplotlib.pyplot as plt

# Set a seed for reproducibility
np.random.seed(42)

# Given values
observed_proportion = 49 / 80  # observed proportion
null_proportion = 0.5  # null hypothesis proportion
n = 80  # sample size
iterations = 10000  # number of bootstrap samples

# Generate bootstrap samples under the null hypothesis
bootstrap_samples = np.random.binomial(n, null_proportion, iterations) / n

# Calculate the 95% confidence interval
confidence_interval = np.percentile(bootstrap_samples, [2.5, 97.5])

# Calculate p-value (proportion of bootstrap samples greater than or equal to observed proportion)
p_value_bootstrap = np.mean(bootstrap_samples >= observed_proportion)

# Visualization: Histogram of bootstrap samples
plt.figure(figsize=(10, 6))
plt.hist(bootstrap_samples, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(observed_proportion, color='red', linestyle='--', label=f'Observed Proportion ({observed_proportion:.4f})')
plt.axvline(confidence_interval[0], color='green', linestyle='--', label=f'95% CI Lower Bound ({confidence_interval[0]:.4f})')
plt.axvline(confidence_interval[1], color='green', linestyle='--', label=f'95% CI Upper Bound ({confidence_interval[1]:.4f})')

# Add labels and title
plt.title('Bootstrap Distribution of Proportions', fontsize=14)
plt.xlabel('Proportion', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()

# Show plot
plt.show()

# Print results
print(f"95% Confidence Interval: {confidence_interval}")
print(f"Bootstrap p-value: {p_value_bootstrap}")

```

Based on the bootstrap analysis, we found the following:
The 95% confidence interval for the proportion of students who can correctly identify the pouring order is between 0.3875 and 0.6125. This interval contains values below 0.5 (random guessing), but it also reaches as high as the observed proportion of 0.6125, suggesting 95% confident that the whole STA130 group would lie on this range.
The bootstrap p-value is 0.0294, which is less than the common significance level of 0.05.
Given that the p-value is below 0.05, we reject the null hypothesis, which suggests that students in STA130 are able to identify the pouring order at a rate significantly better than random chance. However, the confidence interval shows that while the observed success rate is 0.6125, some uncertainty remains about the true proportion.
