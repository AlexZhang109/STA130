```python
import pandas as pd

url = "https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.csv"
# fail https://github.com/KeithGalli/pandas/blob/master/pokemon_data.csv
pokeaman = pd.read_csv(url) 
pokeaman

import statsmodels.formula.api as smf

model1_spec = smf.ols(formula='HP ~ Q("Sp. Def") + C(Generation)', data=pokeaman)
model2_spec = smf.ols(formula='HP ~ Q("Sp. Def") + C(Generation) + Q("Sp. Def"):C(Generation)', data=pokeaman)
model2_spec = smf.ols(formula='HP ~ Q("Sp. Def") * C(Generation)', data=pokeaman)

model2_fit = model2_spec.fit()
model2_fit.summary()
```

R-squared measures the proportion of the variation in the outcome variable explained by the model as a whole. In this case, an R-squared of 17.6% implies that the model explains only a small portion of the variability in the outcome. This does not necessarily mean the model is poor in all respects; it simply indicates that other unexplained factors account for most of the outcome's variation. 
On the other hand,  p-values shows the strength of evidence against the null hypothesis that the coefficients are zero (no effect). A low p-value for a coefficient suggests strong evidence that this independent variable has a significant relationship with the outcome variable, assuming other predictors are not changing.
For categorical predictors like "Generation," Using "Generation 1" as a baseline and make other generations as binary indicators allows each generation level to have its own effect on the outcome, without implying a continuous orrelationship.
